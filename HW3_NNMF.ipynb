## NMF

Non-negative Matrix Factorization is a model where a matrix V is factorized into two matrices W and H, with the property that all three matrices have no negative elements. This non-negativity makes the resulting matrices easier to interpret.

Fill in the following class that implements a NMF by multiplicative updates using the Frobenius norm or the Kullback-Leiber divergence as loss function (implement both). Try 10 different random initializations and choose the best one.

```python
class MyNMF():
    
    def __init__(self, n_components, loss, epsilon = 1e-4, max_iter = 200):
        '''
        Attributes:
        
        n_components_ : integer
            the unknown dimension of W and H (N, r, p)
        max_iter_: integer
            maximum number of iterations
        epsilon_: float
            convergence
        loss_ = {"Frobenius", "KL"}
        w_: np.array
            W Matrix factor
        H_: np.array
            H Matrix factor
        '''
        self.n_components_ = n_components
        self.max_iter_ = max_iter
        self.loss_ = loss
        self.epsilon_ = epsilon
        self.W_ = None
        self.H_ = None
        
    def fit_transform(self, X):
        """ Find the factor matrices W and H
        
        Parameters:
        -----------
        X: (n, p) np.array
            Data matrix
        
        Returns:
        -----
        self
        """        
        # TODO:
        # initialize both matrices
        #            random(0, 1)

        # While not(convergence):
        #     Update W
        #     Update H
        
        # Return self
        
        N = self.n_components_[0]
        r = self.n_components_[1]
        p = self.n_components_[2]
        
        # initialization of W_best and H_best
        tmp_W_best = np.zeros((N,r))
        tmp_H_best = np.zeros((r,p))
        
        for m in range(10):
            
            # initialization of W and H
            tmp_W = np.random.rand(N,r)
            tmp_H = np.random.rand(r,p)
            
            # initialization of W_kl and H_kl
            W_kl = np.zeros(N,r)
            H_kl = np.zeros(r,p)
            
            n=0
            convergence = False
                
            while(n<self.max_iter_ and not(convergence)):
                    
                if(self.loss_ == "Frobenius"):
                    # updating tmp_H
                    num = tmp_W.T.dot(X)
                    den = tmp_W.T.dot(tmp_W.dot(tmp_H))
                    
                    tmp_H = np.multiply(tmp_H, np.divide(num, den))
                    
                    # updating tmp_W
                    num = X.dot(tmp_H.T)
                    den = tmp_W.dot(tmp_H.dot(tmp_H.T))
                    
                    tmp_W = np.multiply(tmp_W, np.divide(num, den)) 
                    
                elif(self.loss_ == "Kullback-Leiber"):
                    
                    W_kl = tmp_W
                    H_kl = tmp_H
                    
                    # updating tmp_W
                    for k in range(r):
                        den = np.sum(H_kl[k,:])
                        H_replicate = np.repeat(H_kl[k,:].reshape((1,p)), N, axis=0)
                        num = np.sum( np.divide(np.multiply(H_replicate,X), W_kl.dot(H_kl)), axis = 1)
                        
                        W_kl[:,k] = tmp_W[:,k] * num / den
                    
                    tmp_W = W_kl
                    
                    # updating tmp_H
                    for k in range(r):
                        den = np.sum(W_kl[:,k])
                        W_replicate = np.repeat(W_kl[:,k].reshape((N,1)), p, axis=1)
                        num = np.sum( np.divide(np.multiply(W_replicate,X), W_kl.dot(H_kl)), axis = 0)
                        
                        H_kl[k,:] = tmp_H[k,:] * num / den
                        
                    tmp_H = H_kl
                       
                # updating convergence
                if(np.linalg.norm(X - tmp_W.dot(tmp_H)) < self.epsilon_):
                    convergence = True
                        
                n = n + 1
                
                #print(n)
                                
            # updating tmp_W_best and tmp_H_best
            if(np.linalg.norm(X - tmp_W.dot(tmp_H)) < np.linalg.norm(X - tmp_W_best.dot(tmp_H_best))):
                tmp_W_best = tmp_W
                tmp_H_best = tmp_H 
            
            print(m)
                                        
        # updating self.W_ and self.H_
        self.W_ = tmp_W_best
        self.H_ = tmp_H_best
```

### First application

For the first application you are going to analyse the following data to give an interpretation of the factorization:

```python
from sklearn.datasets import fetch_olivetti_faces

dataset = fetch_olivetti_faces(shuffle=True)

faces = dataset.data
image_shape = (64, 64)

n_samples, n_features = faces.shape

def plot_faces(title, images, image_shape, n_col=5, n_row=5, cmap=plt.cm.gray):
    plt.figure(figsize=(2. * n_col, 2.26 * n_row))
    plt.suptitle(title, size=16)
    for i, comp in enumerate(images):
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
        plt.imshow(comp.reshape(image_shape), cmap=cmap,
                   interpolation='nearest',
                   vmin=-vmax, vmax=vmax)
        plt.xticks(())
        plt.yticks(())
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)
    
plot_faces("Some faces", faces[:25], image_shape)

faces.shape
```

### Second application

Import the 20newsgroups dataset (from sklearn.datasets import fetch_20newsgroups_vectorized) that contains a collection of ~18,000 newsgroup documents from 20 different newsgroups.

Model the topics present in a subsample with NMF. Print the most common words of each topic.

```python
# TODO

from sklearn.datasets import fetch_20newsgroups 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# number of topics
r = 20

data = fetch_20newsgroups(subset="all", remove=('headers', 'footers', 'quotes')).data
 
# convert the text to a tf-idf weighted term-document matrix 
vectorizer = TfidfVectorizer(max_features=2500, min_df=10, stop_words='english')
 
X = vectorizer.fit_transform(data)
 
index_to_word = np.array(vectorizer.get_feature_names())
 
# apply NMF
nmf = NMF(n_components=r)
 
W = nmf.fit_transform(X)
 
H = nmf.components_
 
# print the most common words for eache topic
for i, topic in enumerate(H):
    print("Topic {}: {}".format(i + 1, ",".join([str(x) for x in index_to_word[topic.argsort()[-10:]]])))
```

The H matrix contains the weigth corresponding to each feature.
